**Download and Load the Data**

### Task 1: Visualizing Data Composition with Pie Charts


##### 1.1 Create a Pie Chart of the Top 5 Databases Respondents Want to Work With


In the survey data, the `DatabaseWantToWorkWith` column lists the databases that respondents wish to work with. Let’s visualize the top 5 most-desired databases in a pie chart.



```python
##Write your code here
all_dbs = df['DatabaseWantToWorkWith'].dropna().str.split(';').explode()
top_5_dbs = all_dbs.value_counts().head()

plt.figure(figsize=(8, 8))
plt.pie(
    top_5_dbs.values,
    labels=top_5_dbs.index,
    autopct='%1.1f%%',
    startangle=90
)

plt.title('Top 5 Most Desired Databases')
plt.axis('equal')  # Equal aspect ratio ensures the pie chart is circular.
plt.show()
```

The `DevType` column lists the developer types for respondents. We’ll examine the distribution by showing the top 5 developer roles in a pie chart.



```python
##Write your code here
all_dev = df['DevType'].dropna().str.split(';').explode()
top_5_dev = all_dev.value_counts().head()

plt.figure(figsize=(8, 8))
plt.pie(
    top_5_dev.values,
    labels=top_5_dev.index,
    autopct='%1.1f%%',
    startangle=90,
    colors= sns.color_palette("Greens")
)

plt.title('Top 5 developer roles', y=1.11)
plt.axis('equal')  
plt.show()
```

##### 1.3 Create a pie chart for the operating systems used by respondents for professional use


The `OpSysProfessional` use column shows the operating systems developers use professionally. Let’s visualize the distribution of the top operating systems in a pie chart.



```python
##Write your code here
all_os = df['OpSysProfessional use'].dropna().str.split(';').explode()
top_10_os = all_os.value_counts().head(10)

sns.set_style("whitegrid")
plt.figure(figsize=(8, 8))
plt.pie(
    top_10_os.values,
    labels=top_10_os.index,
    autopct='%1.1f%%',
    startangle=90,
    colors=sns.color_palette("BuPu")
)

plt.title('Top 10 operating systems used by respondents for professional use', y=1.11)
plt.axis('equal')  
plt.show()
```
